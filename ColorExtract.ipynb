```python
import imageio
import cv2
import glob

for filename in glob.glob('./images/*'):
    print(filename)
    im = cv2.imread(filename, 1)
```

```python
print(im)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

rgbitems
uniset = set()
colors = []
for row in im:
    for col in row:
        uniset.add(tuple(np.array(col).tolist()))
        rgbitems.append(col)
        colors.append('#%02x%02x%02x' % (col[2],col[1],col[0]))
        #
```

```python
print(im.shape)
print(len(rgbitems))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for sp in rgbitems:
    ax.scatter(sp[2], sp[1], sp[0], c='#%02x%02x%02x' % (sp[2], sp[1], sp[0]), marker='o')
    
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')

plt.show()
```
